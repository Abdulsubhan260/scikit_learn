```python
# splitting data into train and test
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
X ,y=load_iris(return_X_y=True)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
# len(X_train)
# len(X_test)
# len(y_test)
len(y_train)
# by using this we can splitt our datsets into training and testing portions.we can mention portion for trainingand testing it will randomly
# distribute our datasets into that particular potions.
```

```python
# in some cases when we split our datasets it will randomly split data so in some cases may be in our train/test portion we might have one component
# of data is way more than as compared to other.this can effect our results in training and evaluating our model.to overcome this problem we simple use
# "stratified_shuffle_split" because this will overcome our problem.
# we just need to visualize our data to make sure it need stratified_shuffle_split or not.

import numpy as np
import matplotlib.pyplot as plt
data=load_iris()
X,y=data.data,data.target
counts=np.bincount(y_train)
positions=np.arange(3)
plt.bar(positions,counts)
plt.xticks(positions,data.target_names)
plt.ylabel("count")
plt.show()

# as we can see in graph that "verginica" portion is lesser than as compare to others and it can effecr our results that why we
# we use stratified_shuffle_split

```

```python
counts=np.bincount(y_test)
positions=np.arange(3)
plt.bar(positions,counts)
plt.xticks(positions,data.target_names)
plt.ylabel("count")
plt.show()
```

```python
# using stratifiedshufflesplit
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index,test_index in split.split(X,y):
    X_train,X_test=X[train_index],X[test_index]
    y_train,y_test=y[train_index],y[test_index]
```

```python
counts=np.bincount(y_train)
positions=np.arange(3)
plt.bar(positions,counts)
plt.xticks(positions,data.target_names)
plt.ylabel("count")
plt.show()
```

```python
counts=np.bincount(y_test)
positions=np.arange(3)
plt.bar(positions,counts)
plt.xticks(positions,data.target_names)
plt.ylabel("count")
plt.show()
```

we can see before and after results of stratifiedshufflesplit as after both

train and test data have similar and equal portions for all three species from dataset.

SCALING DATA :

Scaling data means transforming your dataset so that the numerical features fall within a similar range.

It is a preprocessing step in Machine Learning and Data Analysis to make sure that no single feature dominates others just because it has larger numerical values.

mean = 0

standard deviation = 1

```python
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
X ,y=load_iris(return_X_y=True)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
X_train_scaled
```

```python
# as we can see some values are in negative form in upper scaled data and if we want our data to be postive and scaled between o and 1 we can use
# "minmaxscaler" this will press data between 0 and 1 and in positive form.

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
X_train_scaled

# as we can see now data is scaled between o and 1.
```

```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import OrdinalEncoder
data=fetch_openml('car',as_frame=True).frame
data

```

## OrdinalEncoder:

it is used to convert categorical data into numerical values beacsue most of the

machine learning algorithms learn data in numerical form


in lower given code i selct 2 columns from car data set that are in categorical form(low,med,high) and convert them into 0  1 and 2 because it is easy to use

in machine learning algorithms

0 for low


1 for med

2 for high/big

```python
columns_to_encode=['lug_boot','safety']
encoder=OrdinalEncoder(
    categories=[['small','med','big'],
                ['low','med','high']
                ]
)
data[columns_to_encode]=encoder.fit_transform(data[columns_to_encode])
data
```

## OneHotEncoder:
we use OneHotEncoder to turn categories without any natural order (like colors or cities) into numbers that computers can understand. This avoids confusing our machine learning model by making it think there an arbitrary "bigger" or "smaller" relationship between those categories.


it is used to convert a whole column into 0 and 1 .1 for yes and 0 for no.

in below code i select two columns (occupation and race) drop them and then concate them into new columns by using pandas.



```python
from sklearn.preprocessing import OneHotEncoder
data=fetch_openml('adult',as_frame=True).frame
data.occupation.value_counts()
```

```python
import pandas as pd
encoder=OneHotEncoder(handle_unknown='ignore',sparse_output=False)
encoded_value=encoder.fit_transform(data[['occupation','race']])
new_cols=encoder.get_feature_names_out(['occupation','race'])
# encoded_value
new_cols
```

```python
df_encoded=pd.DataFrame(encoded_value,columns=new_cols,index=data.index)
data_final=pd.concat(
    [data.drop(columns=['occupation','race']),df_encoded],axis=1

)
data_final
```
