## Stratifiedshufflesplit():


this function(stratifiedshufflesplit) is used in the case when the distribution in training data and testing data is not same.before using this function the graph bar of all species(setosa,versicolor and virginica) are not same because they are randomly distributed in both training and testing data set but after using this function we can see both training and testing data sets are equally balanced.

```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1,test_size=0.2)
for train_idx,test_idx,in split.split(X,y):
  x_train,x_test=X[train_idx],X[test_idx]
  y_train,y_test=y[train_idx],y[test_idx]
```
